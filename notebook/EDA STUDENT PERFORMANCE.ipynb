```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Select only numeric columns for aggregation
numeric_columns = ['math score', 'reading score', 'writing score', 'total score', 'average']
numeric_df = df[numeric_columns]

# Group by 'parental level of education' and calculate the mean
grouped_df = numeric_df.groupby(df['parental level of education']).mean()

# Plot the mean values
grouped_df.plot(kind='barh', figsize=(10, 10))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Mean Score')
plt.title('Mean Scores by Parental Level of Education')
plt.show()
```
